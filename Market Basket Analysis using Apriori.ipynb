```python
df1.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
df1 = df1[~df1['InvoiceNo'].str.contains('C')]
```

```python
df1['Country'].value_counts()
```

```python
# Seperating transactions for spain
basket = (df1[df1['Country']=='Spain']
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
country = df1.iloc[:,-1].unique()
country = list(country)
country
```

```python

for i in range(1,6):
    # Seperating transactions for spain
    basket1 = (df1[df1['Country']== country[i]]
               .groupby(['InvoiceNo', 'Description'])['Quantity']
               .sum().unstack().reset_index().fillna(0)
               .set_index('InvoiceNo'))
    print(basket1)
```

```python
basket
```

```python
# Converting all positive values to 1 and everything else to 0

def encode_data(a):
    if a <= 0:
        return 0
    if a >= 1:
        return 1
basket_sets = basket.applymap(encode_data)
basket_sets.drop('POSTAGE', inplace=True, axis=1)  # Removing 'POSTAGE' as an item
```

# Training the Model

```python
# Generating frequent itemsets

frequent_item = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
# Generating rules

rules = association_rules(frequent_item, metric='lift', min_threshold=1)
```
